```python
import tiktoken

def count_tokens(text, model="text-embedding-3-small"):
    encoding = tiktoken.encoding_for_model(model)
    return len(encoding.encode(text))

count_tokens(token_str)
```

```python
def trim_to_token_limit(text, max_tokens=8000, model="text-embedding-3-small"):
    encoding = tiktoken.encoding_for_model(model)
    tokens = encoding.encode(text)

    if len(tokens) <= max_tokens:
        return text  # Already within limit

    # Decode only the first `max_tokens` tokens back to text
    trimmed_text = encoding.decode(tokens[:max_tokens])
    
    return trimmed_text

token_str_trimmed = trim_to_token_limit(token_str, 8000)
count_tokens(token_str_trimmed)
```
