```python
import os
import re
import json
import utils

from dotenv import load_dotenv


load_dotenv(".env")


from embedding import OpenAIEmbeddingModel  # noqa
from vectorindex import vector_search_index  # noqa

embedding_model = OpenAIEmbeddingModel(
    api_key=os.getenv("OPENAI_API_KEY"),
    embedding_endpoint=os.getenv("OPENAI_API_ENDPOINT_EMBEDDINGS")
)
os.getenv("OPENAI_API_ENDPOINT_EMBEDDINGS")
```

```python
search_term = "How can you contact customer support for the X-Flight Drone?"
max_result_count = 5

embeddings = embedding_model.embed(search_term)[0]

query = vector_search_index.generate_query(
    embeddings=embeddings,
    max_result_count=max_result_count
)

results = vector_search_index.search(query)
for result in results:
    print(result)
```
