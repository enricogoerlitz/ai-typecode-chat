```python
import os
import json

from dotenv import load_dotenv
load_dotenv("./.env")

from etl import utils
```

```python
df_catalog = utils.get_imt_catalog()
df_catalog.head()
```

```python
folder = "./etl/data/02-transform-content"
files = [file for file in os.listdir(folder) if file.endswith(".txt")]

for file in files:
    filepath = os.path.join(folder, file)
    print(filepath)

    with open(filepath, "r") as f:
        doc = json.loads(f.readline())
        content = f.read()

    df_result = df_catalog[df_catalog["DokName"] == file.replace(".txt", "")]
    type_codes = [code for code in df_result["Typcode"].fillna("").astype(str).unique() if code != ""]  # noqa
    doc["typeCodes"] = type_codes

    doc_json = json.dumps(doc)
    final_content = "\n\n".join([doc_json, content])

    with open(filepath, "w") as f:
        f.write(final_content)
```
