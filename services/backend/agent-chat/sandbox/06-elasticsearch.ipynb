```python
from elasticsearch_dsl import Document, DenseVector, Keyword, Text, Integer, connections

_ = connections.create_connection(hosts=["http://localhost:9200"])


class DocumentIndex(Document):
    id = Keyword(required=True)
    deviceID = Keyword()
    deviceTypeID = Keyword()
    documentID = Keyword()
    documentName = Text(analyzer='standard')
    documentPageNumber = Integer()
    documentPageContent = Text(analyzer='standard')
    documentPageContentEmbedding = DenseVector(dims=3072)
    metadata_json = Text()

    class Index:
        name = 'device-type-agent-index-001'

DocumentIndex.init()  # creates the index, of not exists 

doc = DocumentIndex(
    id="helloworld2",
    deviceID="dummy",
    deviceTypeID="dummy",
    documentID="dummy",
    documentName="dummy",
    documentPageNumber=1,
    documentPageContent="dummy",
    documentPageContentEmbedding=[0.1] * 3072,
    metadata_json="dummy"
)

doc.save()
DocumentIndex.search().count()
```

```python
doc.search().filter("term", id="helloworld").execute()
```

```python
from elasticsearch_dsl import Search

vector = [0.1] * 3072
s = Search(index="device-type-agent-index-001")

s = s.knn(
    field="documentPageContentEmbedding",
    k=5,
    num_candidates=10,
    query_vector=vector
)

s.execute()
```
