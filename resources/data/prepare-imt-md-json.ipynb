```python
import pandas as pd
```

```python
XL_FILES_PATH = "./imt/files.xlsx"
XL_DEVICE_TYPES_PATH = "./imt/imt_deviceTypes.xlsx"

df_files = pd.read_excel(XL_FILES_PATH)[["Typcode"]]
df_device_types = pd.read_excel(XL_DEVICE_TYPES_PATH)
```

```python
typecodes = list(df_files["Typcode"].drop_duplicates())
df_device_types = df_device_types[df_device_types["Typcode"].isin(typecodes)]

df_device_types.shape, len(typecodes)
```

```python
df_device_types.head()
```

```python
umlaut_map = {
    "ä": "ae",
    "ö": "oe",
    "ü": "ue",
    "ß": "ss",
    "-": "_",
    "(": "",
    ")": "",
    ".": "",
    "/": "_",
    "[": "",
    "]": "",
    " ": ""  # Remove spaces
}

def clean_column_name(col):
    for key, value in umlaut_map.items():
        col = col.replace(key, value)
    return col

# Apply to all columns
df_device_types.columns = [clean_column_name(col) for col in df_device_types.columns]

# Print new column names
print(df_device_types.columns)
```

```python
df_device_types = df_device_types.fillna("")
df_device_types
```

```python

data = df_device_types.to_dict(orient="records")

data
```

```python
import json

with open("../../services/frontend/agent-platform/src/assets/deviceTypesData.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4, default=str, ensure_ascii=False)
```
